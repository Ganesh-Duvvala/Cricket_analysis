```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
links = list()
def espn_webscraping(url):
    page1 = requests.get(url)
    soup = BeautifulSoup(page1.text,'html')
    theadtag = soup.find('thead')
    trtag = theadtag.find_all('tr')
    for tdtag in trtag:
        td_Tag = tdtag.find_all("td")
        td_data = [td_text.text for td_text in td_Tag]
    df = pd.DataFrame(columns = td_data)
    table = soup.find("tbody")
    table_data = table.find_all('tr')
    for row in table_data:
        row_data = row.find_all('td')
        Rows = [row_text.text.strip() for row_text in row_data]
        for a in row_data:
            i = a.find_all('a')
            link = [j['href'] for j in i]
        links.append(link)
        length = len(df)
        df.loc[length] = Rows
    return df
espn_webscraping('https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338')
```

```python
url_match = []
pages=[]
for i in links:
    url1 = str(i)
    url_matchs = 'https://www.espncricinfo.com//'+url1[3:-2]
    url_match.append(url_matchs)

for i in range(len(links)):
    page = requests.get(url_match[i])
    pages.append(page)

for i in range(len(pages)):
    soup = BeautifulSoup(pages[i].text,'html')
    def function(n):
        theadtag = soup.find_all('thead')[n]
        trtag = theadtag.find_all('tr')
        for tdtag in trtag:
            td_Tag = tdtag.find_all("th")
            td_data = [td_text.text for td_text in td_Tag]
            #df_batting_header = pd.DataFrame(columns = td_data)
        Rows_data = []
        table = soup.find_all("tbody")[n]
        table_data = table.find_all('tr')
        for row in table_data:
            row_data = row.find_all('td')
            Rows = [row_text.text.strip() for row_text in row_data]
            Rows_data.append(Rows)
        return Rows_data,td_data
    for j in range(4):
        Rows_data,td_data = function(j)
        if j%2 ==0:
            df_batting = pd.DataFrame(Rows_data,columns = td_data)
            with pd.ExcelWriter("Batting1.xlsx",mode = 'a',if_sheet_exists='new',engine='openpyxl') as writer1:
                df_batting.to_excel(writer1)
        else:
            df_Bowling = pd.DataFrame(Rows_data,columns = td_data)
            with pd.ExcelWriter("Bowling1.xlsx",mode = 'a',if_sheet_exists='new',engine='openpyxl') as writer1:
                df_Bowling.to_excel(writer1)
```

```python
def modify(Workbook):
    list_of_df = []
    df_bt_workbook = pd.ExcelFile(Workbook)
    df_bt_sheetname = df_bt_workbook.sheet_names
    for items in df_bt_sheetname:
        df=df_bt_workbook.parse(items)
        list_of_df.append(df)
    data = pd.concat(list_of_df,ignore_index=True)
    if Workbook == "Batting1.xlsx":
        data = data.drop(columns="Dismissal").set_index("BATTING").dropna(thresh=6)
        data.rename(columns={data.columns[0]:"Dismissal"},inplace = True)
        data["OUT/NOTOUT"] = data.Dismissal.apply(lambda x: "out" if len(x)>=8 else x)
        data.drop(columns="Dismissal",inplace=True)
    else:
        data = data.drop(columns="Unnamed: 0").set_index("BOWLING").dropna(thresh=9)
    return data
data = modify("Batting1.xlsx")
```
